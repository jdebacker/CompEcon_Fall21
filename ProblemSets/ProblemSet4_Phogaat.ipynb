## Clean the data
### Select only male heads of household who are between 25 and 60 years of age and earn wages > $7/hr.

```python
ds.shape
```

```python
# Check if there are any rows which have hannhrs = 0
len(ds[ds.hannhrs == 0])
```

```python
# We drop rows with hannhrs = 0 to avoid dividing denominator by 0
ds = ds[ds.hannhrs != 0]
ds.shape
```

```python
# Adding a new column for wage of individuals
ds["w"] = ds["hlabinc"]/ds["hannhrs"]
```

```python
ds1 = ds[(ds["hsex"] == 1) & (ds["age"].between(25, 60)) & (ds["w"] > 7)]
ds1.shape
```

```python
ds1[["hsex","age","w"]].describe()
```

### Create indicator and continuous variables as necessary

```python
# Checking all the race and the count of each race in our dataset
ds1.groupby(["hrace"]).size()

# Given information about race of the head:
# 1 = White
# 2 = Black
# 3 = Native American
# 4 = Asian/Pacific
# 5 = Hispanic
# 6,7 = Other
```

```python
race = pd.get_dummies(ds1["hrace"])
race.head(n=5)
```

```python
# Renaming the dummy column names
race = race.rename(columns = {1.0:"White",2.0:"Black",3.0:"Other"})
race.head(n=5)
```

```python
# Combining the dummy variables with the dataset
ds2 = pd.concat([ds1,race], axis=1)
ds2.shape
```

```python
# Creating age square variable 
ds2["age_sq"] = ds2["age"]**2

# Changing wage to ln(wage)
ds2["ln_w"] = np.log(ds2["w"])

ds2.shape
```

```python
ds2[["hyrsed","hsex","age","age_sq","w","ln_w","White","Black","Other"]].describe()
```

### Estimate the model via a Maximum Likelihood Estimator separately for t = 1971, 1980, 1990, 2000

```python
# Keeping only the essential columns needed to estimate the model
final = ds2[["ln_w","hyrsed","age","age_sq","Black","Other","year"]]

# Dropping rows with null values
final = final.dropna()

final.shape
```

```python
# Estimating the parameters using Maximum Likelihood Estimator for given years

data = {}
y = {}
result = {}
time = [1971, 1980, 1990, 2000]
initial_value = np.array([0, 0.05, 0.06, -0.02, -0.05, 0])

for i in time:
    data[i] = final[final.year == i]
    y[i] = data[i]['ln_w']
    
    #Objective function
    def objective(param):
        '''
        Args:
        param = coefficient values of the model
        
        Returns:
        e = Sum of squared error term
        neg_LL = Negative likelihood
        
        '''
        beta0 = param[0]
        beta1 = param[1]
        beta2 = param[2]
        beta3 = param[3]
        beta4 = param[4]
        beta5 = param[5]
        
        #predicting value for y (represented as yhat)
        yhat = beta0 + beta1*data[i]['hyrsed'] + beta2*data[i]['age'] + beta3*data[i]['age_sq'] + beta4*data[i]['Black'] + beta5*data[i]['Other']
        
        #sum of squared error term -> summation of (actual value of y - predicted value of y)
        e = np.array((y[i] - yhat)**2).sum()
        return e
        
        #Negative likelihood function estimation
        neg_LL = -np.sum(stats.norm.logpdf(y[i], loc=yhat[i]))
        return(neg_LL)
    
    #Optimal minimized function
    result[i] = opt.minimize(objective, initial_value, method='L-BFGS-B')
    print('Year: ',i ,'\nBeta 1' ,result[i].x[1])
```

### Estimate the model via an ordinary least square estimator

```python
# Ordinary Least Square estimation results for each year
data = {}
x = {}
result = {}
time = [1971, 1980, 1990, 2000]

for i in time:
    data[i] = final[final.year == i]
    x[i] = data[i][["hyrsed","age","age_sq","Black","Other"]]
    result[i] = sm.OLS(data[i]["ln_w"], x[i], missing = 'drop').fit()
    print('Year: ',i ,'\n' ,result[i].params)
```

#### The beta1 value tells us the change in wage associated with one-unit change of the education variable, all other parameters being held constant.
#### The parameter estimates for the given model shows that there is an increase in wage with each given year (1971, 1980, 1990, 2000)
#### We see an increase in wage using both MLE and OLS estaimation methods.
